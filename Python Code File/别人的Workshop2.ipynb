# Section 2: Data Wrangling

### 1. Read in the meteorite data from the Meteorite_Landings.csv file:

```python
meteorite = pd.read_csv('data/Meteorite_Landings.csv')
meteorite
```

### 2. Rename the mass (g) column to mass , and drop all the latitude and longitude columns.

```python
meteorite = meteorite.rename(columns = {'mass (g)': 'mass'})
meteorite = meteorite.drop(['reclat', 'reclong', 'GeoLocation'], axis=1)
meteorite
```

### 3. Update the year column to only contain the year, and create a new column indicating if the year is unknown. Hint: Use year.str.slice() to grab a substring.

```python
meteorite['year'] = meteorite['year'].str.slice(start=6, stop=10)
meteorite['y'] = meteorite['year'].isnull()
meteorite
```

### 6. Create a pivot table that shows both the number of meteorites and the 95th percentile of meteorite mass for those that were found versus observed falling per year from 1990 to 2000 (inclusive).

```python
def percentile_95(x):
    
    return x.quantile(0.95)

summary_func = {'mass': [percentile_95, 'count']}
t6 = meteorite.copy()
t6 = t6[(t6['year']>='1990') & (t6['year']<='2000')]
t6.pivot_table(index='year', columns='fall', values='mass', aggfunc=summary_func)
```

### 2. Using seaborn , create a heatmap that shows the 2019 TSA median traveler throughput by day of week and month.

```python
import seaborn as sns
import numpy as np

tsa_passenger['day of week'] = pd.to_datetime(tsa_passenger['Date']).dt.dayofweek
tsa_passenger['day of month'] = pd.to_datetime(tsa_passenger['Date']).dt.day

t2 = tsa_passenger.pivot_table(index='day of month', columns='day of week', values='2019 Traveler Throughput', aggfunc=np.median)
sns.heatmap(t2)
```

### 3. Annotate the medians in the box plot from #1. Hint: The x coordinates will be 1, 2, and 3 for 2019, 2020, and 2021, respectively. Alternatively, to avoid hardcoding values, you can use the Axes.get_xticklabels() method, in which case you should look at the documentation for the Text class.

```python
import matplotlib.pyplot as plt
from matplotlib.cbook import boxplot_stats

t3 = tsa_passenger[['2019 Traveler Throughput', '2020 Traveler Throughput', '2021 Traveler Throughput']]
box_plot = t3.plot(kind='box', figsize=(10, 8))
#plt.xticks(range(1, 4), [1, 2, 3])
t3_stats = [boxplot_stats(t3[col].dropna().values)[0] for col in t3.columns]
stats = pd.DataFrame(t3_stats, index=t3.columns).iloc[:, [8]]

# annotate
for xtick in box_plot.get_xticks():
    for col in stats.columns:
        box_plot.text(xtick, stats[col][xtick-1], stats[col][xtick-1], horizontalalignment='left', size='small', color='k', weight='semibold', bbox=dict(facecolor='lightgray'))
```

#### Exercises

##### 1. Read in the data and convert the column names to lowercase to make them easier to work with.

```python
import pandas as pd
flight_data = pd.read_csv('data/T100_MARKET_ALL_CARRIER.zip')
flight_data.columns = flight_data.columns.str.lower()
flight_data
```

##### 2. What columns are in the data?

```python
flight_data.columns
```

##### 3. How many distinct carrier names are in the dataset?

```python
len(flight_data['carrier_name'].unique())
```

##### 4. Calculate the totals of the `distance`, `freight`, `mail`, and `passengers` columns for flights from the United Kingdom to the United States.

```python
uk_to_us = flight_data[(flight_data['origin_country_name']=='United Kingdom') 
                       & (flight_data['dest_country_name']=='United States')]
uk_to_us[['distance', 'freight', 'mail', 'passengers']].sum()
```

##### 5. Find the top 5 carriers for distance traveled.

```python
flight_data.groupby('unique_carrier_name')['distance'].sum().sort_values(ascending=False).head(5)
```

##### 6. Find the total cargo transported (mail + freight) and distance traveled for the 10 carriers that transported the most cargo.

```python
t6 = flight_data.copy()
t6['cargo'] = t6['mail'] + t6['freight']
t6.groupby('unique_carrier_name')[['cargo', 'distance']].sum().sort_values(by='cargo', ascending=False).head(10)
```

##### 7. Which 10 carriers flew the most passengers out of the United States to another country?

```python
t7 = flight_data[(flight_data['origin_country_name']=='United States') 
                 & (flight_data['dest_country_name']!='United States')].copy()
carriers_7 = t7.groupby('unique_carrier_name')['passengers'].sum().sort_values(ascending=False).head(10)
carriers_7
```

##### 8. For each of the carriers found in *#7*, find the most popular destination country outside of the United States.

```python
t8 = t7[t7['unique_carrier_name'].isin(carriers_7.index)]
pop_dest = t8.groupby('unique_carrier_name')['dest_country_name'].agg(lambda x:x.value_counts().index[0])
pop_dest
```

##### 9. For each of the carriers found in *#7*, find the total number of passengers flown on international flights to/from the destinations in *#8* or the United States. Note that this dataset only has data for flights with an origin and/or destination of the United States.

```python
def filter_func(x):
    if (x['origin_country_name'] in pop_dest.values) or (x['dest_country_name'] in pop_dest.values):
        return True
    return False

t9 = t8[t8.apply(filter_func, axis=1)]
t9.groupby('unique_carrier_name')['passengers'].sum()
```

##### 10. Between which two cities were the most passengers flown? Make sure to account for both directions.

```python
def cities_func(x):
    if x['origin_city_name'] <= x['dest_city_name']:
        return x['origin_city_name'] + str(' and ') + x['dest_city_name']
    else:
        return x['dest_city_name'] + str(' and ') + x['origin_city_name']

t10 = flight_data.copy()
t10['cities'] = t10.apply(cities_func, axis=1)
pop_cities = t10.groupby('cities')['passengers'].sum().sort_values(ascending=False).head(1)
pop_cities
```

##### 11. Find the top 3 carriers for the pair of cities found in *#10* and calculate the percentage of passengers each accounted for.

```python
pop_cities_name = pop_cities.index[0]
t11 = t10[t10['cities']==pop_cities_name]
carriers_11 = t11.groupby('unique_carrier_name')['passengers'].sum().sort_values(ascending=False)
carriers_11 = (carriers_11 / carriers_11.sum()).head(3)
carriers_11
```

##### 12. Find the percentage of international travel per country using total passengers on class F flights.

```python
def other_countries_func(x):
    if x['origin_country_name'] == 'United States':
        return x['dest_country_name']
    else:
        return x['origin_country_name']


t12 = flight_data[(flight_data['class'] == 'F') & (flight_data['origin_country'] != flight_data['dest_country'])].copy()
t12 = t12.groupby('origin_country_name')['passengers'].sum().sort_values(ascending=False)
t12 = (t12 / t12.sum())
t12
```

##### 13. Using a crosstab, find the percentage of total passengers on class F international flights between US cities and the countries found in *#12* that used the carriers found in *#11*. 

```python
def us_cities_func(x):
    if x['origin_country_name'] == 'United States':
        return x['origin_city_name']
    else:
        return x['dest_city_name']

def other_countries_func(x):
    if x['origin_country_name'] == 'United States':
        return x['dest_country_name']
    else:
        return x['origin_country_name']

    

t13 = flight_data[(flight_data['class'] == 'F') & (flight_data['origin_country'] != flight_data['dest_country'])].copy()
t13 = t13[t13['unique_carrier_name'].isin(carriers_11.index)]
t13['us cities'] = t13.apply(us_cities_func, axis=1)
t13['other countries'] = t13.apply(other_countries_func, axis=1)

ct = pd.crosstab(t13['us cities'], t13['other countries'], values=t13['passengers'], aggfunc=np.sum)
ct / ct.sum()
```

##### 14. Create a pivot table showing the total passengers transported between cities in the United States and other countries by the carriers identified in *#7*. Select the top 10 US cities and top 10 international countries from the result.

```python
t14 = flight_data[flight_data['origin_country'] != flight_data['dest_country']].copy()
t14 = t14[t14['unique_carrier_name'].isin(carriers_7.index)]
t14['us cities'] = t14.apply(us_cities_func, axis=1)
t14['other countries'] = t14.apply(other_countries_func, axis=1)
pivot_14 = t14.pivot_table(index='us cities', columns='other countries', values='passengers', aggfunc=np.sum)
pivot_14
```

```python
pivot_14.sum(axis=1).sort_values(ascending=False).head(10)
```

```python
pivot_14.sum(axis=0).sort_values(ascending=False).head(10)
```

##### 15: For the top 15 international countries, find the percentage of class F passengers traveling to/from the top 10 US cities for international travel (e.g. if only cities A, B, and C flew into Aruba, the sum of the Aruba row/column would be 1). Plot the result as a heatmap.

```python
top_10_us_cities = pivot_14.sum(axis=1).sort_values(ascending=False).head(10).index
top_15_inter_countries = pivot_14.sum(axis=0).sort_values(ascending=False).head(15).index

t15 = flight_data[(flight_data['class'] == 'F') & (flight_data['origin_country'] != flight_data['dest_country'])].copy()
t15['us cities'] = t15.apply(us_cities_func, axis=1)
t15['international countries'] = t15.apply(other_countries_func, axis=1)
t15 = t15[(t15['us cities'].isin(top_10_us_cities)) & (t15['international countries'].isin(top_15_inter_countries))]
t15 = t15.pivot_table(index='us cities', columns='international countries', values='passengers', aggfunc=np.sum)

a = t15.sum(axis=1).copy()
p = t15.div(a, axis='rows').copy()
sns.heatmap(p)
```
