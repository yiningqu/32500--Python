````python
def is_palindrome(string):
    list1 = []
    list2 = []
    for i in range(len(string)):
        list1.append(string[i])
    for j in range(len(string)-1,-1,-1):
        list2.append(string[j])
    if list1 == list2:
        return True
    else:
        return False




#test code
string1 = "radar"
string2 = "python"
print(is_palindrome(string1))  # Output: True
print(is_palindrome(string2))  # Output: False
````

````python
#!/bin/python3

import math
import os
import random
import re
import sys
import hashlib
import pandas
import numpy


'''
@param STRING json_string
@return PANDAS.DATAFRAME
@requirements
- Construct and return a Pandas Dataframe 
from the provided string of JSON formmatted data.
- Do not modify data, data types, names, or indicies.
'''
def load_data_from(json_string):
    import pandas as pd
    return pd.read_json(json_string)

    
'''
@param STRING json_string
@return PANDAS.DATAFRAME
@requirements
- Construct and return a Pandas Dataframe (df). HINT: use 'load_data_from'
- df indexed on 'date' from JSON data
- df has one column for each unique value in 'feature' from JSON data
- df populated with 'value' from JSON data
'''
def tabulate(json_string, pivot_cols=['feature']):
    df = load_data_from(json_string)
    df = df.pivot(index = 'date', columns = pivot_cols, values = 'value')
    df.columns.name = None
    return df
    

    
'''
@param STRING json_string
@return PANDAS.DATAFRAME
@requirements
- Construct and return a Pandas Dataframe (df)
- df contains only the rows for 'feature' from tabulated JSON data where
absolute percent change in 'feature' is greater than the 30-day 
rolling standard deviation of 'feature'
''' 
def analyze1(json_string, feature='Volume'):
    df = tabulate(json_string, pivot_cols=['feature'])
    df = df[feature][df[feature].pct_change().abs() > df[feature].pct_change().rolling(30).std()]
    return df
    
    
'''
@param STRING json_string
@return PANDAS.DATAFRAME
@requirements
- Construct and return a Pandas Dataframe (df)
- df is a statistical summary of 'High' minus 'Low' prices grouped by day of week
    (i.e. 0=Monday, 1=Tuesday, ..., 6=Sunday)
- statistics of interest are mean, standard deviation, 50th and 95th quantiles
- df is sorted in descending order by the mean
'''
def analyze2(json_string, feature_a='High', feature_b='Low'):
    import pandas as pd
    df = tabulate(json_string, pivot_cols='feature')
    #change date to week
    df.index = pd.to_datetime(df.index)
    df.index = df.index.dayofweek
    df['difference'] = df[feature_a] - df[feature_b]
    df_mean = df['difference'].groupby(df.index).mean()
    df_std = df['difference'].groupby(df.index).std()
    df_q50 = df['difference'].groupby(df.index).quantile(0.5)
    df_q95 = df['difference'].groupby(df.index).quantile(0.95)
    #sorting dataframe
    new_df = pd.DataFrame({'mean': df_mean,'std': df_std,'q50': df_q50,'q95': df_q95}).sort_values(by='mean', ascending=False)

    return new_df
    
````

Complete the class Fraction by giving it two public member variables: numerator and denominator. 

Override the addition, subtraction, multiplication, and division functions so that these mathematical operators mimic how fractions work. You must also reduce the fraction.

````python
from math import gcd

class Fraction:
    def __init__(self, numerator, denominator):
        common_factor = gcd(numerator, denominator)
        self.numerator = numerator // common_factor
        # // means return integer instead of float
        self.denominator = denominator // common_factor
    
    def __add__(a, b):
        numerator = a.numerator * b.denominator + b.numerator * a.denominator
        denominator = a.denominator * b.denominator
        common_factor = gcd(numerator, denominator)
        return Fraction(numerator//common_factor, denominator//common_factor)
    
    
    def __sub__(a,b):
        numerator = a.numerator * b.denominator - b.numerator * a.denominator
        denominator = a.denominator * b.denominator
        common_factor = gcd(numerator, denominator)
        return Fraction(numerator//common_factor, denominator//common_factor)
        
        
    def __mul__(a,b):
        numerator = a.numerator * b.numerator
        denominator = a.denominator * b.denominator
        common_factor = gcd(numerator, denominator)
        return Fraction(numerator//common_factor, denominator//common_factor)
    
    def __truediv__(a,b):
        numerator = a.numerator * b.denominator
        denominator = a.denominator * b.numerator
        common_factor = gcd(numerator, denominator)
        return Fraction(numerator//common_factor, denominator//common_factor)
````

Use polymorphism to avoid having to redefine any unnecessary functions. All of these classes must support the base class functions, but it is just as important to not have redundant code. The goal is to write as little code as possible, which means identifying which functions are necessarily to implement and which functions are able to be used from the base class.

````python
import math

class Shape:
    #当明明设了initial variable仍然显示missing positional arguments，给每个variable加上一个initial value就好了
    def __init__(self, height=0, width=0):
        self.__height = height
        self.__width = width
        
    def getDimensions(self):
        return self.__height, self.__width
    
    def setDimensions(self, height, width):
        self.__height = height
        self.__width = width
        
    def calcArea(self):
        raise NotImplementedError

class Rectangle(Shape):
    def calcArea(self):
        height, width = self.getDimensions()
        return height * width

class Square(Rectangle):
    def setDimensions(self, height):
        super().setDimensions(height,height)

class Oval(Shape):
    def calcArea(self):
        height, width = self.getDimensions()
        return math.pi * height * width
    
class Circle(Oval):
    def setDimensions(self, height):
        super().setDimensions(height,height)
````

````python
from abc import ABC, abstractmethod

#这里需要括号里面加ABC
class ExchangeStock(ABC):
    def __init__(self, price):
        self._price = price
    
    
    #记住abstractmethod怎么写
    @abstractmethod
    def getPrice(self):
        pass

    @abstractmethod
    def getMarketCap(self):
        pass

class NasdaqStock(ExchangeStock):
    #如果有missing positional argument，设initial value
    def __init__(self, price, shares=0):
        super().__init__(price)
        self.__shares = shares
    
    def getPrice(self):
        return self._price
    
    def setShares(self, shares):
        self.__shares = shares
    
    def getShares(self):
        return self.__shares
    
    def getMarketCap(self):
        return self._price * self.__shares

class DowStock(ExchangeStock):
    def __init__(self, price, divisor=0):
        super().__init__(price)
        self.__divisor = divisor
    
    def getPrice(self):
        return self._price
        
    def getDivisor(self):
        return self.__divisor
    
    def setDivisor(self, divisor):
        self.__divisor = divisor
    
    def getMarketCap(self):
        return self._price * self.__divisor

#可以inherit from multiple classes
class Apple(NasdaqStock, DowStock):
    def __init__(self, price, shares, divisor):
        #如果是两个super classes，则需要identify哪个，不能用self.super()了
        NasdaqStock.__init__(self, price, shares)
        DowStock.__init__(self, price, divisor)
````

````python
def permutator(s, r):
    if r == 0:
        yield ''
        return
    
    for i in range(len(s)):
        for sub_perm in permutator(s[:i] + s[i+1:], r-1):
            yield s[i] + sub_perm
            
'''
eg. for ('abcd',3)
yield a + sub_perm['bcd',2]
yield b + sub_perm['cd',1] then yield c + sub_perm['bd',1]
yield c then yield d
so first is abc
so second is abd
so third is acb
so forth is acd
'''
part = permutator("abcd", 3)
i = 0
while i<24:
    print(next(part))
    i += 1
````

````python
#using thread 
import threading 
import time

def print_numbers():
    for i in range(1,6):
        print(i)
        time.sleep(1)
        
def print_letters():
    for letter in ['a','b','c','d','e']:
        print(letter)
        time.sleep(1)

t1 = threading.Thread(target=print_numbers)
t2 = threading.Thread(target=print_letters)

# 可以同时run three codes: 
#1. execution line process (python code)
#2. t1: print 1 2 3 4 5 
#3. t2: print a b c d e

#start the threads
t1.start()
t2.start()

#wait for the threads，back to process
t1.join()
t2.join()

print('Done')
#注意：这里的顺序是random的(两个thread同时进行但是产生的顺序是random的)
````

````python
from threading import Lock, Semaphore
from concurrent.futures import ThreadPoolExecutor
import time
import sys
import numpy as np

class Database:

  def __init__(self, quota, limit):
    self.__balance = 0
    self.__deposits = 0
    self.__withdrawals = 0
    self.__open_transactions = 0
    self.__quota = quota
    self.__limit = limit
    self.__mtx = Lock()

  def __repr__(self):
    return f"balance: {self.__balance}\ndeposits: {self.__deposits}\nwithdrawals: {self.__withdrawals}"

  def register_transaction(self, is_deposit:bool):
    with self.__mtx:
      self.__quota -= 1
      if self.__quota < 0:
        raise RuntimeError("too many registrations")
    self.__mtx.acquire()
    if self.__open_transactions + 1 <= self.__limit:
      self.__open_transactions += 1
      self.__mtx.release()
      time.sleep(0.25)
      self.__mtx.acquire()
      if is_deposit:
        self.__deposits += 1
      else:
        self.__withdrawals += 1
      self.__open_transactions -= 1
      self.__mtx.release()
      return True
    else:
      self.__mtx.release()
      return False

  def process_deposit(self, amount):
    balance = self.__balance
    time.sleep(0.0625)      
    balance += amount
    self.__balance = balance
  
  def process_withdrawal(self, amount):
    balance = self.__balance
    time.sleep(0.0625)
    balance -= amount
    self.__balance = balance
    
################################################################################
````

````python
class BinaryTree(Tree):
    def __init__(self, node: Node = None):
        """
        Initialize tree
        hint: an empty tree has a None root
        """
        self.root = node

    def find(self, val: int):
        """
        Locate node equal to val
        hint: this function is not test directly,
            but may be useful in other methods 
            if it returns some kind of node whether 
            found or not
        """
        cur_node = self.root
        while cur_node:
            if cur_node.value == val:
                return cur_node
            elif val < cur_node.value:
                cur_node = cur_node.leftNode
            else:
                cur_node = cur_node.rightNode
        return None
  
    def insert(self, val: int):
        """
        Insert val as a node into tree
        
        hint: do not reinsert an existing node in tree
        """
        if not self.root:
            self.root = Node(val)
        else:
            self._insert(val, self.root)

    def _insert(self, val: int, node: Node):
        if val < node.value:
            if node.leftNode:
                self._insert(val, node.leftNode)
            else:
                node.leftNode = Node(val, aboveNode=node)
                node.update_height()
        elif val > node.value:
            if node.rightNode:
                self._insert(val, node.rightNode)
            else:
                node.rightNode = Node(val, aboveNode=node)
                node.update_height()
      
    def delete(self, val: int):
        """
        Delete node from tree equal to val, if it exists
        """
        node_to_delete = self.find(val)
        if node_to_delete:
            self._delete(node_to_delete)
        
    def _delete(self, node: Node):
        if not node.leftNode and not node.rightNode:
            self._delete_leaf(node)
        elif node.leftNode and node.rightNode:
            self._delete_with_two_child(node)
        elif not node.leftNode or not node.rightNode:
            self._delete_with_one_child(node)
        
            
    def _delete_leaf(self, node: Node):
        parent = node.aboveNode
        if parent:
            if parent.leftNode and parent.leftNode == node:
                parent.leftNode = None
            elif parent.rightNode:
                parent.rightNode = None
            parent.update_height()
        else:
            self.root = None
            
    def _delete_with_one_child(self, node: Node):
        parent = node.aboveNode
        if node.leftNode:
            child = node.leftNode
        else:
            child = node.rightNode
        if parent:
            if parent.leftNode and parent.leftNode == node:
                parent.leftNode = child
            elif parent.rightNode:
                parent.rightNode = child
            child.aboveNode = parent
            parent.update_height()
        else:
            self.root = child
            child.aboveNode = None
            
    def _delete_with_two_child(self, node: Node):
        cur_node = node.rightNode 
        while cur_node.leftNode:  #find leftmost in the right tree把leftmost的值代入变成新的node
            cur_node = cur_node.leftNode
        val = cur_node.value
        self._delete(cur_node)  # delete leftmost one 
        node.value = val   #将先的node的值更新成leftmost的值
        node.update_height()
````

# AVL Tree
- binary search tree 的balanced版本

````python
class AVLTree(Tree):
    def __init__(self, node: Node = None):
        """
        Initialize tree
        
        hint: an empty tree has a None root
        hint: it may be useful to derive and override 
        from your BinaryTree implementaiton
        """
        self.root = node

    def find(self, val: int):
        """
        Locate node equal to val
        hint: this function is not test directly,
            but may be useful in other methods 
            if it returns some kind of node whether 
            found or not
        """
        cur_node = self.root
        while cur_node:
            if cur_node.value == val:
                return cur_node
            elif val < cur_node.value:
                cur_node = cur_node.leftNode
            else:
                cur_node = cur_node.rightNode
        return None
        
    def insert(self, val):
        """
        Insert val as a node into tree
        
        hint: this code is not test directly,
        but may be useful in other methods
        if it returns some kind of node whether
        it is found or not
        """
        # print("insert", val)
        self.root = self._insert(self.root, val)
    
    def _insert(self, node, val):
        if not node:
            return Node(val)
        if val < node.value:
            node.leftNode = self._insert(node.leftNode, val)
            node.leftNode.aboveNode = node
        elif val > node.value:
            node.rightNode = self._insert(node.rightNode, val)
            node.rightNode.aboveNode = node
        else:
            return node
        node = self._balance(node)
        node.update_height()
        return node
        
    def _balance(self, node):
        balance = node.determine_balance()
        if balance > 1: # left heavy
            if node.leftNode.determine_balance() < 0:
                node.leftNode = self._left_rotate(node.leftNode)
            return self._balance(self._right_rotate(node))
        elif balance < -1: # right heavy
            if node.rightNode.determine_balance() > 0:
                node.rightNode = self._right_rotate(node.rightNode)
            return self._balance(self._left_rotate(node))
        else:
            return node
            
    def _left_rotate(self, a):
        above = a.aboveNode
        b = a.rightNode
        c = b.leftNode
        b.leftNode = a
        a.rightNode = c
        a.aboveNode = b
        if c:
            c.aboveNode = a
            b.aboveNode = above
            c.update_height()
        else:
            b.aboveNode = above
            a.update_height()
        return b

    def _right_rotate(self, a):
        above = a.aboveNode
        b = a.leftNode
        c = b.rightNode
        b.rightNode = a
        a.leftNode = c
        a.aboveNode = b
        if c:
            c.aboveNode = a
            b.aboveNode = above
            c.update_height()
        else:
            b.aboveNode = above
            a.update_height()
        return b

    def delete(self, val):
        """
        Delete node equal to val from tree
        """
        # print("delete", val)
        self.root = self._delete(self.root, val)
        
    def _delete(self, node, val):
        if not node:
            return node
        if val < node.value:
            node.leftNode = self._delete(node.leftNode, val)
        elif val > node.value:
            node.rightNode = self._delete(node.rightNode, val)
        else:
            if not node.leftNode and not node.rightNode:
                return self._delete_leaf(node)
            elif node.leftNode and node.rightNode:
                node = self._delete_with_two_child(node)
            else:
                node = self._delete_with_one_child(node)
        
        node = self._balance(node)
        node.update_height()
        return node
            
    def _delete_leaf(self, node: Node):
        parent = node.aboveNode
        if parent:
            if parent.leftNode and parent.leftNode == node:
                parent.leftNode = None
            elif parent.rightNode:
                parent.rightNode = None
            parent.update_height()
        else:
            self.root = None
            
    def _delete_with_one_child(self, node: Node):
        parent = node.aboveNode
        if node.leftNode:
            child = node.leftNode
        else:
            child = node.rightNode
        if parent:
            if parent.leftNode and parent.leftNode == node:
                parent.leftNode = child
            elif parent.rightNode:
                parent.rightNode = child
            child.aboveNode = parent
            parent.update_height()
        else:
            self.root = child
            child.aboveNode = None
        return child
            
    def _delete_with_two_child(self, node: Node):
        cur_node = node.rightNode
        while cur_node.leftNode:
            cur_node = cur_node.leftNode
        val = cur_node.value
        node.value = val
        node.rightNode = self._delete(node.rightNode, val)
        node = self._balance(node)
        node.update_height()
        return node
````

Another aspect of programming applications that involve server connections involves being able to handle messages spanning across multiple packets. For any number of reasons (size of the message, or server-specific settings), messages may be split up across multiple packets. The client must therefore be able to handle data that is split across packets.

- Create an Exception class called MessageFormatException.
- Create a class Receiver that contains a public class member list called messages. Receiver must have two methods:
- read, which takes a string. It must determine if the string contains a full message. If it does, it should add the message to messages. If it does not, it must wait until it gets a full message.
- The string format is such that the first four characters will indicate the length of the message. This will be left-padded with zeroes.
- If the string is not formatted correctly, read should raise a MessageFormatException. - printMessages, which prints the messages in messagesclass

````python
class MessageFormatException(Exception):
    pass

class Receiver:
    def __init__(self):
        self.messages = []
        self.buffer = ""

    def read(self, data):
        self.buffer += data  # Append new data to the buffer

        while True:
            # Check if buffer has enough data for length extraction
            if len(self.buffer) < 4:
                return  # Not enough data to determine message length

            # Extract message length
            try:
                msg_length = int(self.buffer[:4])
            except ValueError:
                raise MessageFormatException("Invalid message format")

            # Check if the entire message is received
            if len(self.buffer) < msg_length:
                return  # Wait for more data

            # Extract and store full message
            self.messages.append(self.buffer[:msg_length])

            # Remove processed message from buffer
            self.buffer = self.buffer[msg_length:]

    def printMessages(self):
        for message in self.messages:
            print(message)
````
